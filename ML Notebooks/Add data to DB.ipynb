```python
from azure.cosmos import CosmosClient
import json
```

```python
# Set up the Cosmos DB client
endpoint_uri = "https://its-project.documents.azure.com:443/"
primary_key = "**"
client = CosmosClient(endpoint_uri, primary_key)
```

```python
# Get a reference to the database and container
database_name = "testsqldb"
container_name = "testpartion1"
database = client.get_database_client(database_name)
container = database.get_container_client(container_name)

```

```python
# Insert data into the container
data = {
    "id": "3",
    "name": "Jayesh Bhatt",
    "age": 25
}
container.create_item(body=data)
```

```python
query = "SELECT * FROM c"
items = list(container.query_items(query, enable_cross_partition_query=True))
```

```python
# Print the retrieved data
for item in items:
    print(item)
```
